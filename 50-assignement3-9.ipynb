```python
def factorial_finder(num:int) ->int:
    """
    Calculate the factorial of a non-negative integer.
    
    Paremeter:
        num (int): The number for which factorial is required.
    Returns:
        factorial of a given number.
    Raises:
        TypeError:if num is not a integer.
        ValueError:if num is negative
    
    """
    if isinstance(num,bool):
        raise TypeError("Num is not a integer")
    if not isinstance(num,int):
        raise TypeError("input must be an integer")
    if num < 0:
        raise ValueError("factorial is not define is negative number")
    factorial = 1
    for i in range(1,num+1):
        factorial = factorial * i
    return factorial
if __name__ == "__main__":
    try:
        number = input("Enter the number to find the factorial :")
        number = int(number)
        # if not isinstance(number,int):
        #     raise ValueError("Input is not a integer please enter a numrica value")
        print("Factorial:",factorial_finder(number))
    except ValueError as e:
        print(e)
    except TypeError as e:
        print("Error:",e)
```

```python
""" Check the string is palindrome."""

def is_palindrome(text: str) -> bool:
    """
    Checks whether a given string is palindrome.
    
    Parameter:
          Text(str):Check the given string is palindrome or not.
    Return:
          Return true if string is palindrome false otherwise.
    Raises:
          TypeError:if input is not a string.
          ValueError:if input string is empty.
    
    """
    # Validate the input is string.
    if not isinstance(text,str):
        raise TypeError("input must be a string")
    # Validate input is not empty
    if text.strip() == "":
        raise ValueError("input string can not be empty")
    else:
        text = text.strip().lower()
    # reverse the string.
    reversed_string = ""
    for ch in text:
        reversed_string = ch + reversed_string
    return text == reversed_string

# Main Exucation
if __name__ == "__main__":
    try:
        user_input = input("Enter the string to check palindrome:")
        if is_palindrome(user_input):
            print(f"{user_input} is a palindrome.")
        else:
            print(f"{user_input} is not a palindrome.")
    except (TypeError,ValueError) as e:
        print("error:",e)
    # except Exception as e:
    #     print(e)
```

```python
def count_vowels(text:str) -> int:
    """
    Count the number of vowels in a given string.

    Parameters:
        text(str): The input string.
    Returns:
        int: Numbers of vowels in a string.
    Raises:
        TypeError:if input is not a string.

    """
    # validate the input is a string.
    if not isinstance(text,str):
        raise TypeError("input must be a string.")
    # Logic
    vowels = "aeiou"
    count = 0
    for char in text.lower():
        if char in vowels:
            count = count + 1
    return count

# Main Exucation
if __name__ == "__main__":
    try:
        user_input = input("Enter the string to count the vowels:")
        result = count_vowels(user_input)
        print(f"Total vowels : {result} ")
    except TypeError as e:
        print(e)
    except Exception as e:
        print(e)
```

```python
from typing import List,Any
def duplicate_remover(List:List[Any]) -> List[Any]:
    """
    Remove duplicate from a list while maintaining original order.

    Parameter:
         List(List(Any)): A list coutaining any type of element.
    Returns:
         List(Any):A new list from duplicate remover.
    Raises:
         TypeError: if input is not a list.
         ValueError: if list ia empty.
    
    """
    # validate the input is a list
    if not isinstance(List,list):
        raise TypeError("input must be a list.")
    # validate list is not empty
    if List == 0:
        raise TypeError("list is not empty.")
    # Validate list has only one argument.
    if List == 1:
        print("list has only one argument: no duplicate to remove")
    result = []
    for i in List:
        if i not in result:
            result.append(i)
    return result
# main Exucation
if __name__ == "__main__":
    number_list = [100,29,300,5,5,4,6,6,1]
    
    try:
        result = duplicate_remover(number_list)
    except (TypeError,ValueError) as error:
        print("error:",error)
    except Exception as error:
        print("error:",error)
    else:
        print("After removing duplicate: ",result)
```

```python
from typing import List
def second_largest(nums:List[int]) -> int:
    """
    Find the second largest in a list without using a sorting.
    
    Parameter:
         nums(List[int]): A list of integer.
    
    Returns:
         int: Second largest in a list.
    
    Raises:
        TypeError:if input is not a list or element not a integer.
        ValueError:if the list has fewer then 2 unique element.
    
    """
    # Validate the input is a list.
    if not isinstance(nums,list):
        raise TypeError("input must be a list.")
    # Validate the all element in a list are integer
    if any(not isinstance(current_number,int) for current_number in nums):
        raise TypeError("All element in a list should be a integer.")
    # check list has atlest two element.
    if len(nums) < 2:
        raise ValueError("input has contain atlest two element.")
    if nums[0] > nums[1]:
        largest = nums[0]
        second = nums[1]
    else:
        largest = nums[1]
        second = nums[0]
    for current_number in nums[2:]:
        if current_number > largest:
            second = largest
            largest = current_number
        elif current_number != largest and current_number > second:
            second = current_number
    if second == largest:
        raise ValueError("list must contain atlest two unique alement.")
    return second

# ------------  Main Exucation (entry point) -----------
if __name__ == "__main__":
    try:
        numbers_list = [51,15,67]
        result = second_largest(numbers_list)
    except (TypeError,ValueError) as error:
        print(f"error: {error}")
    except Exception as e:
        print(e)
    else:
        print("Second largest :",result)
```
