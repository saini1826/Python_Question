```python
def reverse_string(input_string: str) -> str:
    """
    Function to reverse a string using for loop.
    Args:
       Input_str(str) The string to be reversed.
    Return:
       Str: the reversed version of a string.
    Raises:
       TypeError: If input not a string.
    
    """
    if not isinstance(input_string,str):
        raise TypeError("input must be a string")
    reversed_str = ""
    for char in input_string:
        reversed_str = char + reversed_str
    return reversed_str
if __name__ == "__main__":
    try:
        # Taking input from the user
        user_input = input("Enter the string to reverse: ")
        # calling the function
        result = reverse_string(user_input)
        # display the reverse string
        print(f"Reverse string :{result}")
    except TypeError as e:
        print(e)
    except Exception as e:
        print(e)
        
```

```python
def find_largest_number(numbers_list:list[float]) -> float:
    """
    function return a largest number in a list.
    Args:
        number_list(number_list:list[float]): A List is taking a numeric value like(int or float).
    Return:
        float: A largest number in a list.

    Raises:
        TypeError: If input is not a list its contain a non numeric value.
        ValueError: if the list is empty.
        
    """
    # Validate is input is a list
    if not isinstance(numbers_list,list):
        raise TypeError("input must be a list type of float or int")
    # Validate that list is not empty
    if not numbers_list:
        raise ValueError("list can not be empty")
    # validate that list element should be numeric
    for num in numbers_list:
        if not isinstance(num,(int,float)):
            raise TypeError("All element must be numeric")
    largest = numbers_list[0]
    for num in numbers_list:
        if num > largest:
            largest = num
    return largest
# Main exucation
if __name__ == "__main__":

    try:
        #taking input from the user
        user_input = input("Enter number in a list seprate by space int or float only: ")
        # Converting user_input in a list
        numbers_list = [float(item)for item in user_input.split()]
        # Calling the function
        result = find_largest_number(numbers_list)
        #Display the result
        print("Largest number in the list is :",result)

    except (TypeError,ValueError) as error:
        print(error)
    except Exception as e:
        print(e)
    
    
            
    
```

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    print(n)
    
    return n * factorial(n-1)    
```

```python
def my_factorial(number:int) -> int:
    """
    Function to find the factorial of a number.
    
    Parameter:
          number(int): factorial of a given number.
    
    Return:
         int: Factorial of a number.
    raises:
         TypeError:if input is not a integer.
         ValueError:if input is a non negative integer.  
    
    """
    if not isinstance(number,int):
        raise TypeError("input must be an integer.")
    if number < 0:
        raise TypeError("Input must be a non negative integer.")
    # factorial 
    if number == 0 or number == 1:
        return 1
    return number * my_factorial(number - 1)

####### Main (Exucation) ########
if __name__ == "__main__":
    num = 5
    try:
        result = my_factorial(num)
        print(f"factorial of {num} is: {result}")
    except (TypeError,ValueError) as error:
        print("error:",error)
    except Exception as e:
        print(e)
```

```python
""" function to conut the charcater in a string."""

def count_characters(input_string:str) -> dict:
    """
    Function to count the occorrences of each character in a string.
    
    Args:
       input_string(str): The string whose charecter to be counted.
    
    Returns:
       dict: A dictonary with character as key and their count as value.
    Raises:
       TypeError: if input is not a string.
    
    """
    if not isinstance(input_string,str):
        raise TypeError("Input must be a string.")
    char_count = {}
    for char in input_string:
        if char in char_count:
            char_count[char] += 1
        else:
            char_count[char] = 1
    return char_count

####### Main(Euxcation) #######
if __name__ == "__main__":
    user_input = "Apple"
    try:
        result = count_characters(user_input)
        print(result)
    except (TypeError,ValueError) as error:
        print("error:",error)
    except Exception as e:
        print(e)
```
