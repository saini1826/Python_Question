```python
def find_median(numbers:list) ->float:
    """
    Calculate the median of a list of numbers.

    Parameters
    ----------
    numbers : list
        A list of numeric values (int or float).

    Returns
    -------
    float
        The median value of the list.

    Raises
    ------
    TypeError
        If the input is not a list or contains non-numeric values.
    ValueError
        If the list is empty.
    
    Examples
    --------
    >>> find_median([1, 2, 3, 4, 5])
    3.0
    >>> find_median([1, 2, 3, 4])
    2.5
    """
    # Exception handling
    if not isinstance(numbers, list):
        raise TypeError("Input must be a list.")
    if len(numbers) == 0:
        raise ValueError("List is empty. Cannot compute median.")
    if not all(isinstance(x, (int, float)) for x in numbers):
        raise TypeError("All elements in the list must be numeric.")

    numbers = len(nums)

    if numbers % 2 != 0:
        median = nums[numbers//2]
    else:
        median = (nums[numbers//2 - 1] + nums[numbers//2]) / 2
    
    return median
     
##### main(exucation) ######
if __name__ == "__main__":
    nums = [4,3,6,7,8]
    try:
        result = find_median(nums)
        print("median:",result) 
    except (TypeError,ValueError) as error:
        print("error:",error)
    except Exception as e:
        print(e)
```

```python
def sort_dict_by_value(data,reverse=False):
    """
    Return a sort dictionary by a value.
    
    Parameter
    --------
    Data : dict: dictionary to be sorted by a value.value of a pair of key:value.
    
    Return
    ------
    Dict: A dictionary to be sorted by a value.

    Raises
    ------
    TypeError: if input is not a dict.
    ValueError: if input is an empty.
    
    """
    if not isinstance(data,dict):
        raise TypeError("input must be a dict.")
    if len(data) == 0:
        raise ValueError("Dictionary is empty. cannot sort")
    
    sorted_items = sorted(data.items(), key=lambda item: item[1], reverse=reverse)
    return sorted_items
 
if __name__ == "__main__":
    dict1 = {'apple':5,'banana':2,'jackfruit':8,'cherry':3}
    try:
        result = sort_dict_by_value(dict1)
        print("Dictionary is sorted by a value:",result)
    except (TypeError,ValueError) as error:
        print("error:",error)
    except Exception as e:
        print(e)
    


```

```python
def is_subsequence(large, small):
    """
    Check if a subsequence exists within a given sequence.

    Parameters
    ----------
    subseq : list
        The subsequence to check.
    seq : list
        The main sequence in which to search.

    Returns
    -------
    bool
        True if subseq is a subsequence of seq, False otherwise.

    Raises
    ------
    TypeError
        If inputs are not lists.
    ValueError
        If either subseq or seq is empty.

    Examples
    --------
    >>> is_subsequence([1, 3], [1, 2, 3, 4])
    True
    >>> is_subsequence([2, 5], [1, 2, 3, 4])
    False
    """
    if not isinstance(large, list) or not isinstance(small, list):
        raise TypeError("Both subseq and seq must be lists.")
    if len(large) == 0 or len(small) == 0:
        raise ValueError("Neither subseq nor seq can be empty.")

    j = 0
    for i in large:
        if j < len(small) and i == small[j]:
            j += 1
    return j == len(small)
 
if __name__ == "__main__":
    large = [2,3,4,5,6]
    small = [2,4,9,8,1]
    try:
        result = is_subsequence(large,small)
        print("Dictionary is sorted by a value:",result)
    except (TypeError,ValueError) as error:
        print("error:",error)
    except Exception as e:
        print(e)
```

```python
def common_element_in_list(list1,list2,list3) ->list:
    """
    Function to return a common element in a three list.
    
    Args:
      number(list): Return a list.
      
    Return:
      list: Given a common element in a three list.
    
    Raises:
      TypeError: if input is not a list.
      ValueError: if input is an empty.
    
    """
    if not isinstance(list1,list) or not isinstance(list2,list) or not isinstance(list3,list):
        raise TypeError("input must be a list")
    if len(list1) == 0 or len(list2) == 0 or len(list3) == 0:
        raise TypeError("input is empty cannot given element.")
    # if not all isinstance(list1,list2,list3 (int, float)):
    #     raise TypeError("All elements in the list must be numeric.")
    
    common = []
    for i in list1:
        if i in list2 and i in list3:
            common.append(i)
    return common

if __name__ == "__main__":
    list1 = [23,34,4,3,2]
    list2 = [43,432,4,2]
    list3 = [4,34,44,2]
    try:
        result = common_element_in_list(list1,list2,list3)
        print(" common element in a list:",result)
    except (TypeError,ValueError) as error:
        print("error:",error)
    except Exception as e:
        print(e)
            
```
