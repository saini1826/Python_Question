```python
def upper_maker(func):
    def inner_function():
        value = func()
        return value.upper()
    return inner_function

def spliter_maker(func):
    def inner_function():
        value = func()
        return value.split()
    return inner_function

@spliter_maker
@upper_maker
def say_hi():
    return "hello good morning"

say_hi()
```

```python
def globel_fun(func):
    def inner_function():
        value = func().split("@")
        name = value[0].split(".")
        return f"hello {name[0]} {name[1]} how are you"
        return inner_fun

@globel_fun
def email():
    return "abhishek.saini@gmail.com"

email()
```

```python
def logger(func):
    def wrapper(*args, **kwargs):
        print(f"running {func.__name__} with argumwnt {args}")
        result = func(*args, **kwargs)
        print(f"{func.__name__} returned {result}")
        return result
    return wrapper

@logger
def add(a, b):
    return a+b

add(10, 5)
```

```python
# second method hai
my_function = logger(add)
```

```python
my_function(10, 5)
```
