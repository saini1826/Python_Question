# try except else finally

```python
number = int(input("enter number :"))
print(100/number)
```

```python
try:
    number = int(input("enter number :"))
    print(100/number)
except ZeroDivisionError:
    print("don't type zero ")
except Exception as e:
    print(e)

print("hello")
```

```python
try:
    number = int(input("enter number :"))
    print(100/number)
except ZeroDivisionError:
    print("don't type zero ")
except ValueError:
    print("only  number are allow")
except Exception as e:
    print(e)

else:
    print("i am working becouse try is working")

finally:
    print("i am work not matter what happen above")
```

```python
def multiply(x,y):
    return x*y

if __name__ == "__main__":
    try:
        x = int(input("enter number :"))
        y = int(input("enter number :"))
        print(multiply(x,y)) 
    except Exception as e:
        print(e,"only number")
```
