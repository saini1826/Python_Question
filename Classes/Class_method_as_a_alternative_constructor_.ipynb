```python
class myclass:
    def __init__(self, name, city):
        self.name = name
        self.city = city
    @classmethod
    def object_created(cls, list1):
        print("object created")
        return cls(list1[0], list1[1])
```

```python
#  object1 = myclass("abhishek", "noida")
```

```python
new_object_myclass = myclass.object_created(["rahul", "delhi"])
```

```python
isinstance(new_object_myclass, myclass)
```

```python
new_object_myclass.name
```

```python
class myclass:
    def __init__(self, name, city):
        self.name = name
        self.city = city
    @classmethod
    def object_created(cls, string):
        value1 = string.split("@")[0].split(".")[0]
        value2 = string.split("@")[0].split(".")[1][:-5]
        print("object created")
        return cls(value1, value2)
```

```python
new_object_myclass = myclass.object_created("abhishek.saini80610@gmail.com")
```

```python
new_object_myclass.name
```

```python
new_object_myclass.city
```
