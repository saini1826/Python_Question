```python
class A:
    def __init__(self, name):
        self.name = name
    def __str__(self):
        return f"name = {self.name}"

person = A("abhishek")
print(str(person))
```

```python
class A:
    def __init__(self, name):
        self.name = name
    def object1(self):
        return self.name
    def __len__(self):
        return len(self.name)   # ager len chiya to use karne hai
    def __str__(self):
        return f"{self.name}"   # __str__ or __repr__ ya sab dunde method hai
    def __repr__(self):
        return f"n = {self.name}"
```

```python
person = A("abhishek")
```

```python
person.name
print(str(person))
print(repr(person))
print(len(person))
```

```python
class A:
    def __init__(self, name):
        self.name = name
    def object1(self):
        return self.name
    def __add__(self, other):
        return self.name + other.name
    def __len__(self):
        return len(str(self.name))
    def __str__(self):
        return f"{self.name}"   
    def __repr__(self):
        return f"n = {self.name}"
```

```python
a = A(20000)
b = A(30000)
```

```python
print(a+b)
```
