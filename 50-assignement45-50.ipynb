```python
def swap_number(a,b):
    """
    Function to swap to number without a temporary variable.
    
    Parameter:
          number(int): Exchange the value of both variable.
    Returns:
          int: swap two number.
    Raises:
          TypeError: if input is not a integer.
          ValueError: if input is contain non negative integer. 
    """
    try:
        a,b = b,a
        return a, b
    except Exception as e:
        raise ValueError(f"Error while swapping value: {e}")
```

```python
from collections import Counter

def first_non_repeated_char(text: str) -> str:
    """
    Find the first non-repeated character in a string.

    Parameters
    ----------
    s : str
        The input string to be checked.

    Returns
    -------
    str
        The first non-repeated character.

    Raises
    ------
    TypeError
        If the input is not a string.
    ValueError
        If the string is empty or no non-repeated character exists.
    """

    # ✅ Input validation
    if not isinstance(text, str):
        raise TypeError("Input must be a string.")
    if text == "":
        raise ValueError("Input string cannot be empty.")

    # ✅ Count occurrences of each character
    char_count = Counter(text)

    # ✅ Find first non-repeated character
    for char in text:
        if char_count[char] == 1:
            return char

    # # ✅ If no non-repeated character found
    # raise ValueError("No non-repeated character found in the string.")

# main(Execution)
if __name__ == "__main__":
    try:
        user = input("Enter the string :")
        result = first_non_repeated_char(user)
        print(f"{result} is a first non repeated character.")
    except (TypeError,ValueError) as error:
        print("error:",error)
    except Exception as e:
        print(e)

```

```python
def contain_only_digit(number: str) -> bool:
    """
    Check if the input string contains only digits using a for loop.

    Parameters
    ----------
    number : str
        The input string to be checked.

    Returns
    -------
    bool
        True if the string contains only digits, False otherwise.

    Raises
    ------
    TypeError
        If the input is not a string.
    ValueError
        If the string is empty.
    """

    # ✅ Input validation
    if not isinstance(number, str):
        raise TypeError("Input must be a string.")
    if number == "":
        raise ValueError("Input cannot be empty.")

    for num in number:
        if not num.isdigit():
            return False
    return True


# main(Execution)
if __name__ == "__main__":
    try:
        user_input = input("Enter the string: ")
        if contain_only_digit(user_input):
            print(f"{user_input} is a digit.")
        else:
            print(f"{user_input} is not a digit.")
    except (TypeError, ValueError) as error:
        print("Error:", error)
    except Exception as e:
        print(e)
```

```python
### 49 ####
def is_armstrong(number: int) -> bool:
    """
    Check if a given number is an Armstrong number.

    An Armstrong number (narcissistic number) is a number that is equal to the sum
    of its own digits each raised to the power of the number of digits.

    Parameters
    ----------
    number : int
        The integer to be checked.

    Returns
    -------
    bool
        True if the number is an Armstrong number, False otherwise.

    Raises
    ------
    TypeError
        If the input is not an integer.
    ValueError
        If the input is a negative integer (Armstrong numbers are defined for non-negative integers).
    """

    # ✅ Input validation
    if not isinstance(number, int):
        raise TypeError("Input must be an integer.")
    if number < 0:
        raise ValueError("Input must be a non-negative integer.")

    # Convert number to string for digit extraction
    digits = str(number)
    power = len(digits)

    # Calculate Armstrong sum
    total = sum(int(digit) ** power for digit in digits)

    return total == number


# Main execution block
if __name__ == "__main__":
    try:
        user_input = input("Enter a number: ")
        # Convert to integer safely
        num = int(user_input)

        if is_armstrong(num):
            print(f"{num} is an Armstrong number.")
        else:
            print(f"{num} is not an Armstrong number.")

    except ValueError as ve:
        print("Error:", ve)
    except TypeError as te:
        print("Error:", te)
    except Exception as e:
        print("Unexpected error:", e)
```

```python
from typing import List, Any

def generate_subsets(elements: List[Any]) -> List[List[Any]]:
    """
    Generate all subsets (the power set) of a given list.

    Parameters
    ----------
    elements : List[Any]
        The input list whose subsets are to be generated.

    Returns
    -------
    List[List[Any]]
        A list containing all subsets of the input list.
        Each subset is represented as a list.

    Raises
    ------
    TypeError
        If the input is not a list.
    ValueError
        If the input list is empty.
    """

    # ✅ Input validation
    if not isinstance(elements, list):
        raise TypeError("Input must be a list.")
    if len(elements) == 0:
        raise ValueError("Input list cannot be empty.")

    subsets = [[]]  # start with the empty subset

    for element in elements:
        # For each element, add it to existing subsets to form new subsets
        new_subsets = [subset + [element] for subset in subsets]
        subsets.extend(new_subsets)

    return subsets


# Main execution block
if __name__ == "__main__":
    try:
        user_input = input("Enter elements separated by space: ").split()
        result = generate_subsets(user_input)
        print("All subsets (power set):")
        for subset in result:
            print(subset)

    except (TypeError, ValueError) as error:
        print("Error:", error)
    except Exception as e:
        print("Unexpected error:", e)
```
